## Quality Isssues
- Duplicated Records.
- Missing Data.
- Date Data Type.

# Clean

```python
Data_Clean = Data.copy()
```

## Duplicated Data

### Define:
Remove Duplicate Records.

### Code:

```python
Data_Clean.drop_duplicates(inplace = True)
```

### Test:

```python
Data_Clean[Data_Clean.duplicated(keep = False)]
```

```python
Data_Clean.info()
```

## Missing Data

### Define:
Remove The Record that have null.

### Code:

```python
Data_Clean.dropna(inplace = True)
```

### Test:

```python
Data_Clean.isnull().sum()
```

```python
Data_Clean.info()
```

## Inconsistent Data

### Define:
Change Date Date Type.

### Code:

```python
Data_Clean['Accident Date'] = pd.to_datetime(Data_Clean['Accident Date'], format = '%d-%m-%Y')
```

### Test:

```python
Data_Clean.info()
```
